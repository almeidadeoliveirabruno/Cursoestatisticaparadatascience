Agora que você aprendeu o básico sobre correlação, é hora de aplicar esse conhecimento em uma base de dados real! Carregue a base de dados das casas (house-prices.csv) e gere as correlações entre todos os atributos. Quais atributos estão mais relacionados com o preço das casas?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv('house_prices.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset_numeros = dataset.select_dtypes(include=['number'])
dataset_numeros.head()
```

```python
dataset_numeros.corr()
```

```python
plt.figure(figsize=(20,12))
sns.heatmap(dataset_numeros.corr(), annot= True)
```

```python
maiores_cor = dataset_numeros[['price','sqft_lot','grade','sqft_living15']]
```

```python
maiores_cor.corr()
```

```python
sns.scatterplot(data= maiores_cor, x = 'sqft_lot', y = 'price')
```

```python
sns.scatterplot(data= maiores_cor, x='grade', y= 'price')
```

```python
sns.scatterplot(data = maiores_cor, x= 'sqft_living15', y = 'price')
```
