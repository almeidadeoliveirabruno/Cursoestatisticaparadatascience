Agora que você aprendeu os principais tipos de distribuições, o objetivo é utilizar a base de dados do censo para identificar as distribuições para cada um dos atributos. Para isso, carregue o arquivo census.csv e gere o histograma para cada atributo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('census.csv')
dataset.head()
```

```python
col = dataset.columns
print(col[0])
num_col = len(dataset.columns)
print(num_col)
```

```python
sns.displot(dataset['age'], kind = ('hist'), kde = True);
#Gama
```

```python
sns.displot(dataset['final-weight'], kind = ('hist'), kde = True);
#gama
```

```python
sns.displot(dataset['education-num'], kind = ('hist'), kde= True);
#poison ou binomial
```

```python
sns.displot(dataset['capital-gain'], kind = ('hist'), kde = True);
#Exponencial
```

```python
sns.displot(dataset['capital-loos'], kind = ('hist'), kde = True);
#exponencial
```

```python
sns.displot(dataset['marital-status'], kind = ('hist'), kde = True);
#binomial
```

```python
sns.displot(dataset['sex'], kind = ('hist'), kde = True);
#bernoulli
```

```python
sns.displot(dataset['income'], kind = ('hist'), kde = True);
#bernoulli
```
